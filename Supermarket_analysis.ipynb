By inspection, the 'Date' datatype is an object, we need to change it to datetime

#### Feature Engineering (znitelik Mhendislii)

```python
dir(pd) #Learn pandas' code
```

```python
df['Date'] = pd.to_datetime(df['Date']) # change int to datetime type
```

```python
df['date'] = (df['Date']).dt.day
df['month'] = (df['Date']).dt.month
df['year'] = (df['Date']).dt.year
```

```python
df['month_name'] = df['Date'].dt.month_name()
```

```python
df['weekday'] = df['Date'].dt.day_name()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['Hour'] = (df['Time']).dt.hour # type(sales['Time'])
```
